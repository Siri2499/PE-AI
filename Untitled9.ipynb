Compute outer product of 2 vectors

```
import numpy as np
p=[[1,0],[0,1]]
q=[[1,2],[3,4]]
print("original matrix:")
print(p)
print(q)
res=np.outer(p,q)
print("Outer product of two vectros is:")
print(res)
```

Write the peogram to compute the condition of the given original matrix:[[1 2] [3 4]] condition number of the said matrix: 14.9330343737

```
a=[[1,2],[3,4]]
print("original matrix:")
print(a)
c=np.linalg.cond(a)
print("Condition number of the said matrix:")
print(c)
```
